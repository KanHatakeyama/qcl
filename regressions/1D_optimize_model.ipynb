```python

#from qulacsvis import circuit_drawer
from QCLRegressor.regressors.StandardQCLRegressor import StandardQCLRegressor
from QCLRegressor.gates.IsingGates import IsingGates
from QCLRegressor.gates.CNOTRotationGates import CNOTRotationGates
from QCLRegressor.Encoders.XYEncoder import XYEncoder
from QCLRegressor.Encoders.prerocess import *
from QCLRegressor.gates.Observable import ZObservable



x_dim=1
```

```python
from QCLRegressor.regressors.StandardQCLRegressor import StandardQCLRegressor
from QCLRegressor.gates.CNOTRotationGates import CNOTRotationGates
from QCLRegressor.Encoders.XYEncoder import XYEncoder
from QCLRegressor.Encoders.WEncoder import WEncoder
from QCLRegressor.Encoders.prerocess import *
from QCLRegressor.gates.Observable import ZObservable
from QCLRegressor.gates.IsingGates import IsingGates



x_dim=1
n_qubit=2
depth=3
model_dict={}
model_dict["1-XY-XY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["1-XY-XY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["1-XY-XY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["2-X-XY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="x"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["2-X-XY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="x"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=2))
model_dict["2-X-XY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="x"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=3))
model_dict["2-Y-XY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="y"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["2-Y-XY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="y"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=2))
model_dict["2-Y-XY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=WEncoder(n_qubit,angle_mode="y"),vqe_gates=CNOTRotationGates(n_qubit=2,depth=3))
model_dict["2-XY-XY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["2-XY-XY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=2))
model_dict["2-XY-XY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=CNOTRotationGates(n_qubit=2,depth=3))
model_dict["2-XY(a)-XY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit,preprocess_func=two_arc_angles),vqe_gates=CNOTRotationGates(n_qubit=2,depth=1))
model_dict["2-XY(a)-XY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit,preprocess_func=two_arc_angles),vqe_gates=CNOTRotationGates(n_qubit=2,depth=2))
model_dict["2-XY(a)-XY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit,preprocess_func=two_arc_angles),vqe_gates=CNOTRotationGates(n_qubit=2,depth=3))
model_dict["2-XY-IsXY1"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=1))
model_dict["2-XY-IsXY1'"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=1))
model_dict["2-XY-IsXY1''"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=1))
model_dict["2-XY-IsXY2"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=2))
model_dict["2-XY-IsXY2'"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=2))
model_dict["2-XY-IsXY2''"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=2))
model_dict["2-XY-IsXY3"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=3))
model_dict["2-XY-IsXY3'"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=3))
model_dict["2-XY-IsXY3''"]=StandardQCLRegressor(x_dim=x_dim,n_qubit=2,encoder=XYEncoder(n_qubit),vqe_gates=IsingGates(n_qubit=2,depth=3))


n_cols=6
total_models=len(list(model_dict))
n_rows=int(total_models/n_cols)+1

for n in [20]:
    for mode in ["sin","linear"]:
        tr_X,tr_y,te_inner_X,te_inner_y,te_low_X,te_low_y,te_high_X,te_high_y,act_X,act_y=prepare_dataset(n_all_record=n,mode=mode,plot=False)
        i=0

        fig = plt.figure(figsize=(12,2*n_rows),dpi=100)
        for model_name,model in model_dict.items():
            model.fit(tr_X,tr_y)
            plt.subplots_adjust(wspace=0.5, hspace=0.6)
            ax = fig.add_subplot(n_rows, n_cols, i+1)
            eval(ax,model,tr_X,tr_y,te_inner_X,te_inner_y,te_low_X,te_low_y,te_high_X,te_high_y,act_X,act_y,title=model_name)
            i+=1


#circuit_drawer(model.vqe_gates.circuit)
```
