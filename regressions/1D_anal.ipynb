```python

import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

```

# Calculation time

```python
path="res/calc_time_1d_2022-05-11 08:59:35.json" 
path="res/calc_time_1d_2022-05-11 09:16:22.json" 
with open(path,"r") as f:
    json_str=f.read()

res_dict=json.loads(json_str)
res_df=pd.DataFrame.from_dict(res_dict).T
res_df["Qubits"]=res_df["Qubits"].astype(int)
res_df["Depth"]=res_df["Depth"].astype(int)
```

```python

from matplotlib.ticker import MultipleLocator
plt.figure(figsize=(2,2),dpi=150)
sns.lineplot(data=res_df,x="Qubits",y="Time",hue="Depth",err_style="bars")#,ci=None)
plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=8)
plt.ylabel("Training time (s)")
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1)) 
```

```python
plt.figure(figsize=(2,2),dpi=150)
sns.lineplot(data=res_df,x="Depth",y="Time",hue="Qubits",err_style="bars")
plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=8)
plt.ylabel("Training time (s)")
plt.xlabel("Circuit depth")
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1)) 
```
